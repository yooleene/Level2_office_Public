### 9. 빈 값 체크하기
#### 1) isspace 함수
##### ① 문자열의 정보가 모두 공백문자인지를 판단하여 참이면 True, 거짓이면 False로 반환
##### ② 모든 공백문자(White Space: 스페이스, 탭, 개행문자)로 구성되어 있으면 True, 다른 문자가 섞여 있으면 False
##### ③ <span style = 'color : blue'> 스페이스값이 들어간 빈칸까지 False로 체크할 수 있어 엑셀의 ISBLANK보다 더 유용!</span>


```python
df = pd.read_excel('data_set/거래품목.xlsx', index_col = 'id')
df.head()
```

```python
for i in range(len(df['품목'])):
    if str(df['수량'].loc[i]).isspace() == True:
        df._set_value(i, '수량', 0)
    if str(df['단가'].loc[i]).isspace() == True:
        df._set_value(i, '단가', 0)
df = df.astype({'수량' : 'int', '단가' : 'int'})
df['공급가액'] = df['수량'] * df['단가']
df['세액'] = df['공급가액']*0.1
df = df.astype({'세액' : 'int'})
df
```

##### ③ 두 날짜 사이의 차이는 datetime 변수를 연산하여 계산하며, 연산결과는 timedelta 자료형으로 저장됨
- timedelta 자료형은 일, 초, 마이크로초 단위로 저장되므로 산술/대소 연산자를 사용해 계산이 가능

```python
# 1. 일자(day)차이 구하기 + 일자 외 나머지 시간 차이 구하기
gap_days = (now_datetime - this_datetime).days
gap_total_seconds = (now_datetime - this_datetime).seconds
gap_total_hours = gap_total_seconds/3600
gap_total_minutes = gap_total_seconds/60

print('지정시간 :', this_datetime)
print('현재시간 :', now_datetime)
print('일자차이 :', gap_days, '일')
print('차이(일자 + 초 포함) :', gap_days, '일', gap_total_seconds, '초')
print('차이(일자 + 초 → 시간으로 변환하여 포함) :', gap_days, '일', gap_total_hours, '시간')
print('차이(일자 + 초 → 분으로 변환하여 포함) :', gap_days, '일', gap_total_minutes, '분')
```

```python
import math

# 2. 일자, 시간, 분, 초 차이를 구하여 출력하기
gap_hours = math.floor(gap_total_seconds/3600)
gap_minutes = math.floor((gap_total_seconds - (gap_hours * 3600))/60)
gap_seconds = gap_total_seconds - gap_hours*3600 - gap_minutes*60

print('지정시간 :', this_datetime)
print('현재시간 :', now_datetime)
print('시간차이 :',
    str(gap_days) + '일',
    str(gap_hours) + '시간', 
    str(gap_minutes) + '분',
    str(gap_seconds) + '초')
```
