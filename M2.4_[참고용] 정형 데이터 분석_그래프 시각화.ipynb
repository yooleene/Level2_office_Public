### 1. 데이터 빈도/비율 그래프 그리기
#### 1) Pandas plot() 메서드는 matplotlib와 결합해 다양한 시각화 도구를 제공
##### ① 빈도분석을 위한 Bar차트, Pie차트 뿐만 아니라 탐색적 분석을 위한 히스토그램, 산점도 생성이 가능
##### ② matplotlib 패키지와 연결된 pandas plot() 메서드를 통해 pandas에서 직접 그래프 생성 가능

```python
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
```

```python
# 그래프 한글 폰트 적용
mp.rc('font', family = 'Malgun Gothic')

df = pd.read_csv('data_set/학생건강검사 데이터.csv', index_col = 'id')
df
```

#### 2) Bar Chart

### 

```python
# 성별 Bar Chart
s_count = df['sex'].value_counts()  # 개별(1개) 범주형 변수의 자료 수 계산
s_count.plot(kind = 'bar', title = '성별 학생수', figsize = (6, 4), fontsize = 14, color = [(0.3, 0.5, 0.7), (0.7, 0.5, 0.3)], rot = 0)
plt.show()
```

```python
# 도시별 Bar Chart
# groupby(column) : column 기준 자료 요약, as_index : groupby 변수의 index로 사용 여부
c_count = df['city'].value_counts()
c_count.plot(kind = 'bar', title = '지역별 학생수', figsize = (8, 6), fontsize = 16)
plt.show()
```

#### 3) Pie Chart

```python
# 성별 Pie Chart
s_count = df['sex'].value_counts()  # 개별(1개) 범주형 변수의 자료 수 계산
s_count.plot(kind = 'pie', title = '성별 학생수', figsize = (6, 6), fontsize = 14, startangle = 90, autopct = '%.2f%%', explode = (0, 0.05))
plt.show()
```

#### 4) Pandas boxplot()

##### ① 데이터를 4분위로 나눈 다음 중앙값을 중심으로 1~3분위까지 박스를 그려 분포를 표현하는 그래프

```python
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
```

```python
# 그래프 한글폰트 적용
mp.rc('font', family = 'Malgun Gothic')
df = pd.read_csv('data_set/학생건강검사 데이터.csv', index_col = 'id')

columns = ['height', 'weight']
bys = ['sch_grade', 'sex']
for col in columns:
    for by in bys:
        df.boxplot(column = col, by = by, figsize = (6, 4), fontsize = 16)  # 서로 다른 그룹의 by열을 그룹화하는 서로 다른 상자기름을 플로팅 함
        plt.show()
```

#### 5) Pandas hist()

##### ① (연속형) 데이터를 요약하여 가로축에 계급(구간), 세로축에 빈도를 표현하는 히스토그램 그리기

```python
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
```

```python
# 그래프 한글폰트 적용
mp.rc('font', family = 'Malgun Gothic')
df = pd.read_csv('data_set/학생건강검사 데이터.csv', index_col = 'id')
```

```python
# 데이터 전체 히스토그램
df['height'].hist()
plt.show()
```

```python
# 고등학생만 필터링한 히스토그램
s_data = df.loc[df['sch_grade'] == '고']
s_data['height'].hist()
plt.show()
```
